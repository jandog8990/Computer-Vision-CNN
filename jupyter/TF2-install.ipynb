# Anaconda-based Installation
We will try to work with the new release of Tensorflow since it dramatically changes the interface. 

Download <b>Python 3.7 version</b> from: [https://www.anaconda.com/distribution/](https://www.anaconda.com/distribution/).

## Create anaconda environment
After you download and install Anaconda, navigate and find the
<i>Anaconda Prompt</i>. 

Select the <i>Anaconda Prompt</i>, right-click and click on <b>Run as administrator</b>

Inside the <i>Anaconda Prompt</i>, type:
```bash
conda create --name tf2
```

## Install packages in tf2 environment
We will install everything in the tf2 virtual environment.

Open up an <i>Anaconda Prompt</i> <b>as Administrator</b> and type:
```bash
conda activate tf2
conda install pip
pip install tensorflow==2.0.0-beta1

conda install git jupyter spyder sympy matplotlib scikit-learn opencv
conda install pandas
conda install -c anaconda pillow
pip install tensorflow-datasets
conda install pydot graphviz
```

## Remove Anaconda Environment if something goes wrong
Make aure to restart <i>Anaconda Prompt</i> <b>as Administrator</b>.
Then type:
```bash
conda env remove --name tf2
```
You can then restart the installation process. You will get a warning that the tf2 directory exists. This is fine. You can overwrite it.

# Jupyter notebook basics
Saving the notebook does not save the values of Anaconda variables.
It only saves the printed output text. 
You will need to rerun the code to reproduce the same environment with your saved values.

## Start Jupyter notebook WITHOUT Tensorboard
Open up a new <i>Anaconda Prompt</i> and type:
 
    conda activate tf2
    jupyter notebook

## Start Jupyter notebook WITH Tensorboard
Open up a new <i>Anaconda Prompt</i> and type:

    conda activate tf2
    tensorboard --logdir=./my_logs --port=6006

Open a SECOND <i>Anaconda Prompt</i> and type:

    conda activate tf2
    jupyter notebook
    
When done, you can click on <i>Control-C</i> to kill the Tensorboard process.

## Basic operations
<b>Execute a Cell:</b> Click <i>Run</i> or simply hit <i>Control-Enter</i> on the Cell.

<b>Run all from scratch:</b> Click <i>Cell</i> and then <i>Run All</i>.

<b>Save the current notebook:</b> Click on the disk icon to save your work.

## Interrupt long computation
Click on <i>Kernel</i> and then click on <i>Interrupt</i>.
This approach KEEPS ALL VARIABLES.
  
## Restart running after Kernel interrupt
To run the rest do: Click on <i>Cell</i> and select <i>Run All Below</i>.

## Restart up to a point
1. To run up to a point: Click inside the <b>last Cell</b>
2. Click on <i>Kernel</i> and <i>Restart</i> to <b>lose all variables</b>.
3. Click on <i>Cell</i> and <i>Run All Above</i>.

## Exit after all computations are complete
Note: This process will <b>loose all of your variables</b>.
You can only preserve the current outputs.
 1. Click on the disk icon to save your work.
 2. Click <i>Kernel</i> then <i>Shutdown</i>.
 3. Close the browser windows. 
 4. Then close <i>Anaconda Prompt</i>.  


## Emergency exit if Interrupt does not work after long wait
Note: This process will <b>loose all of your variables</b>.
You can only preserve the current outputs.
 1. Save the current notebook by clicking on the disk icon.
 2. Simply type <i>Control-C</i> in the <i>Anaconda Prompt</i>.
 3. You may have to press <i>Control-C</i> multiple times.
 4. You can then click and close each tab in the browser.
